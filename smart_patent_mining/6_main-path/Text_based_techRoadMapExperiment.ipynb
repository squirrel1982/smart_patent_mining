```python
class Technology_life:
    def __init__(self,threshold,tech_path,Text_based_techRoadMap):
        self.threshold = threshold
        self.roadMap = Text_based_techRoadMap
        self.tech_path = tech_path

    def getPatentAboveThresholdByNodeGA(self,nodeGA):
        ab = roadMap.getAbstractByGA(nodeGA)
        result = [nodeGA,[]]
        sims = roadMap.computeSimilarityByTermsAndSort(ab)
#I wanna get a list of patents,which includes patentID，patentGA，similar
#format [PathNodeID，[[patentID，patentGA，similar].......]]
        for i in range(len(sims)):
            if sims[i]>=self.threshold:
                nodeID = roadMap.getNodeIDByNodeGA(self.roadMap.gas[i])
                if nodeID!=None:
                    result[1].append([nodeID,self.roadMap.gas[i],sims[i]])
        return result
    def getPatentAppYearListByGAS(self,gas):
        conn=pymssql.connect(host=".",user="su",password="123",database="car_database")
        cursor = conn.cursor();
        result = []
        sql = "select ga,PI_YEAR from Priority_Info where ga = \'"
        for i in gas:
            cursor.execute(sql+i+"\'")
            m=cursor.fetchall()
            for j in m:
                result.append(j)
        cursor.close()
        conn.commit()
        conn.close()
        return result
    def getAllDataListByGAS(self,gas):
        conn=pymssql.connect(host=".",user="su",password="123",database="car_database")
        cursor = conn.cursor();
        result = []
        sql = " select A.ga,A.code,B.PI_Year from assignee_Info A,[Priority_Info] B  where A.GA = B.GA and A.GA = \'"
        for i in gas:
            cursor.execute(sql+i+"\'")
            m=cursor.fetchall()
            for j in m:
                result.append(j)
        cursor.close()
        conn.commit()
        conn.close()
        return result
    def getPatentAppPersonListByGAS(self,gas):
        conn=pymssql.connect(host=".",user="su",password="123",database="car_database")
        cursor = conn.cursor();
        result = []
        sql = "select ga,code from assignee_Info where ga = \'"
        for i in gas:
            cursor.execute(sql+i+"\'")
            m=cursor.fetchall()
            for j in m:
                result.append(j)
        cursor.close()
        conn.commit()
        conn.close()
        return result
    def getPatentAboveThresholdOfWholePath(self):
        list_1 = []
        for i in self.tech_path:    
            nodeGA = self.roadMap.g.node[i]['ga']
            list_1.append(self.getPatentAboveThresholdByNodeGA(nodeGA))
        return list_1

    def createPatentSet(self,list_1):
        gasFromList_1 = [[j[1] for j in i[1]] for i in list_1]
        gasFromList_1 = sum(gasFromList_1,[])
        gasSet = set(gasFromList_1)
        return gasSet
```
